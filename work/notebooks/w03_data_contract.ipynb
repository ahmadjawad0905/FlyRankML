## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

### Data Limitations

1. **No direct refresh label**
   
   The dataset does not contain a field indicating whether a content page actually needed or received a content refresh. Therefore, the project must rely on a proxy label derived from future performance rather than confirmed editorial decisions.

2. **Incomplete data availability**
   
   Not all clients have both Google Search Console (GSC) and Google Analytics 4 (GA4) data. Some rows only contain GSC metrics, while GA4 metrics are unavailable, limiting the set of features that can be used consistently.

3. **Historical performance only**
   
   The warehouse contains historical search and analytics metrics but does not capture qualitative information such as content quality, writing accuracy, user satisfaction, or competitor changes that may also influence whether a page should be refreshed.

4. **Time-limited snapshot**
   
   This notebook uses only the March 2026 partition, so the analysis reflects the characteristics of that month and does not capture seasonal or long-term trends across the entire dataset.
