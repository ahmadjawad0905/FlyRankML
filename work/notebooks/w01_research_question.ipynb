## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
This project is a decision-support system, not a decision-making system. Its purpose is to identify content pages that appear to deserve review based on observable search and engagement signals. The recommendations are intended to help content teams prioritize their limited review time more effectively.
The results of this project will show associations and patterns found in the data. For example, the model may identify pages with high impressions, low CTR, or declining trends as strong review candidates. These findings suggest that such pages may deserve attention, but they do not prove why the pages performed that way.
This project cannot claim that refreshing a page will increase traffic, that a recommendation will definitely improve performance, or that any observed relationship is causal. Likewise, it cannot predict or explain Google's ranking algorithm or prove that a particular signal directly influences search rankings.
Instead, the model provides evidence-based recommendations that should be reviewed by humans before any action is taken.
```
