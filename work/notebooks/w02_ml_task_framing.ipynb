## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# Classification with ranking output.
# The model will learn whether a content page should be considered a review candidate based on historical search and engagement signals. Rather than producing only a Yes/No decision, it will assign each page a probability of needing review. \
# These probabilities can then be sorted to produce a ranked list of pages, allowing content teams to prioritize their limited review time.
# Although the final output is a ranking, the underlying machine learning task is a binary classification problem ( of whether a page should be reviewed or not).
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
# The dataset does not contain a direct label indicating whether a page should be reviewed. Therefore, I will use a proxy target derived from observed performance trends.
# Initially, pages with a trend_direction of "down" will be treated as label = 1, while pages with stable or improving trends will be labeled as 0.
# This proxy represents a practical approximation of the review decision and can be refined later as I explore additional signals throughout the project.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# The primary success metric for this project will be Precision@K.
# Since the goal is to recommend only a limited number of pages for human review, it is more important that the highest-ranked recommendations are genuinely useful than achieving high overall accuracy.
# A successful model should maximize the proportion of truly valuable review candidates among the top-ranked pages presented to the content team.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# A simple rule such as

# "if (ctr < 0.20) 
#   print("Review candidate")

# ignores many other important factors.

# Whether a page deserves review depends on the interaction of multiple signals, including impressions, clicks, engagement, content age, traffic trends, and historical performance.
# Machine learning can learn complex relationships between these features and assign different levels of importance to each one. This enables more flexible and consistent recommendations than a single manually designed rule.
# Rather than replacing human judgment, the model supports better prioritization by considering many signals simultaneously.
```
